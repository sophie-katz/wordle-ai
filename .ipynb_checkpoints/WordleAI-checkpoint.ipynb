```python
import json
import pandas as pd
```

```python
with open("Words.json", "r") as file:
    words = json.load(file)

print(words[:10])
```

```python
letter_hist_dict = {}

for word in words:
    for letter in word:
        if letter in letter_hist:
            letter_hist_dict[letter] += 1
        else:
            letter_hist_dict[letter] = 1

letter_hist = pd.DataFrame([{"letter": key, "count": value} for key, value in letter_hist_dict.items()])

letter_hist["frequency"] = letter_hist["count"] / letter_hist["count"].max()

letter_hist = letter_hist.sort_values("count", ascending = False)

letter_hist
```

```python
words_by_letter_frequency_sum_data = []

for word in words:
    frequency_sum = letter_hist.loc[letter_hist["letter"].isin(list(word))]["frequency"].sum()
    words_by_letter_frequency_sum_data.append({"word": word, "frequency_sum": frequency_sum})

words_by_letter_frequency_sum = pd.DataFrame(words_by_letter_frequency_sum_data)

words_by_letter_frequency_sum = words_by_letter_frequency_sum.sort_values("frequency_sum", ascending = False)

words_by_letter_frequency_sum
```
